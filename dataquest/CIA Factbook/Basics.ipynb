```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('factbook.db')
q_tables = "SELECT * FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(q_tables, conn)
print(tables)

q_facts = "SELECT * FROM facts LIMIT 5"
five_facts = pd.read_sql_query(q_facts, conn)
print(five_facts)
```

```python
query_mm = '''
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
'''
min_max = pd.read_sql_query(query_mm, conn)
min_max
```

```python
query_ol = '''
SELECT name
FROM facts
WHERE (population = 0) OR (population = 7256490011)
'''
df = pd.read_sql_query(query_ol, conn)
df
```

```python
query_hist = '''
SELECT population, population_growth, birth_rate, death_rate
FROM facts
WHERE name NOT IN ('Antarctica', 'World')
'''
hist_df = pd.read_sql_query(query_hist, conn)
hist_df
```
