```python
import pandas as pd
import numpy as np
import altair as alt
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars.head()
```

```python
yes_no = {
    'Yes':True,
    'No': False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars.head()
```

```python
cols_dict = {'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
        'Unnamed: 4':'seen_2',
        'Unnamed: 5':'seen_3',
        'Unnamed: 6':'seen_4',
        'Unnamed: 7':'seen_5',
        'Unnamed: 8':'seen_6'}
seen_map = {
    "Star Wars: Episode I  The Phantom Menace" : True,
    "Star Wars: Episode II  Attack of the Clones" :True,
    "Star Wars: Episode III  Revenge of the Sith" :True,
    "Star Wars: Episode IV  A New Hope" :True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,
    np.nan : False
}

cols = list(star_wars.columns[3:9])
for col in cols:
    star_wars[col] = star_wars[col].map(seen_map)

star_wars = star_wars.rename(columns=cols_dict)
star_wars.iloc[:,3:9].head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
renames = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'film_1_ranking',
    'Unnamed: 10': 'film_2_ranking',
    'Unnamed: 11': 'film_3_ranking',
    'Unnamed: 12': 'film_4_ranking',
    'Unnamed: 13': 'film_5_ranking',
    'Unnamed: 14': 'film_6_ranking'
}
star_wars = star_wars.rename(columns = renames)
star_wars.iloc[:,9:15]
```
